# Creating Multiple Plots and Subplots in Matplotlib

In this lecture, we’ll learn how to create multiple plots and subplots using **Matplotlib’s object-oriented interface** — the recommended way for clarity and control.

---

## Learning Goals
- Understand the difference between `plt.subplots()` and `plt.subplot()`
- Learn to create subplot grids (1x2, 2x2, etc.)
- Customize each subplot individually
- Share axes across subplots
- Use real-world examples: trigonometric functions and simulated time series

---

## Why Use the Object-Oriented Interface?

Instead of relying on global functions like `plt.plot()`, we use `fig, ax = plt.subplots()` to:
- Create **Figure** and **Axes** objects explicitly
- **Control each subplot** precisely
- Add **labels, limits, titles, and legends** per axis

---

## Dataset 1: Trigonometric Functions

Let’s create subplots for sine, cosine, and tangent waves.


````python
import numpy as np
import matplotlib.pyplot as plt

# Create data
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create a 1x3 grid of subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 4))

# Plot each function on a different subplot
axs[0].plot(x, y1, color='blue')
axs[0].set_title("Sine Wave")
axs[0].set_xlabel("x")
axs[0].set_ylabel("sin(x)")
axs[0].grid(True)

axs[1].plot(x, y2, color='green')
axs[1].set_title("Cosine Wave")
axs[1].set_xlabel("x")
axs[1].set_ylabel("cos(x)")
axs[1].grid(True)

axs[2].plot(x, y3, color='red')
axs[2].set_title("Tangent Wave")
axs[2].set_xlabel("x")
axs[2].set_ylabel("tan(x)")
axs[2].set_ylim(-5, 5)  # Limit y-axis to avoid large values
axs[2].grid(True)

plt.tight_layout()
plt.show()
````

## What Just Happened?

| Element        | Purpose                                 |
|----------------|------------------------------------------|
| `plt.subplots(1, 3)` | Creates a row of 3 axes            |
| `axs[i]`       | Access individual subplot                |
| `set_title()`  | Adds a title to each subplot             |
| `set_ylim()`   | Controls axis limits (e.g. tangent spike)|
| `tight_layout()` | Avoids label overlap                  |

## Dataset 2: Simulated Time Series

Let’s simulate two correlated time series and show:
- Shared x-axis
- Vertical layout (2x1)
- Customized y-axis limits and titles


````python
# Simulated temperature and humidity data
days = np.arange(365)
temperature = 10 * np.sin(2 * np.pi * days / 365) + np.random.normal(0, 2, size=365) + 20
humidity = 50 + 10 * np.cos(2 * np.pi * days / 365) + np.random.normal(0, 5, size=365)

# Create vertically stacked subplots with shared x-axis
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 6), sharex=True)

# Temperature plot
ax1.plot(days, temperature, color='tomato')
ax1.set_title("Daily Temperature Over One Year")
ax1.set_ylabel("Temperature (°C)")
ax1.grid(True)

# Humidity plot
ax2.plot(days, humidity, color='dodgerblue')
ax2.set_title("Daily Humidity Over One Year")
ax2.set_ylabel("Humidity (%)")
ax2.set_xlabel("Day of Year")
ax2.grid(True)

plt.tight_layout()
plt.show()
````

## Key Concepts

- `fig, (ax1, ax2) = plt.subplots(2, 1)` → Creates 2 rows, 1 column layout
- `sharex=True` → Allows zooming or highlighting a region on both plots simultaneously
- Good for time series, comparisons, and sensor data

## Advanced Layout: 2x2 Grid with Titles

Let’s now create a **2x2 grid** showing four different mathematical functions.


````python
x = np.linspace(0, 10, 100)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(x, np.sin(x), label='sin(x)')
axs[0, 0].set_title("Sine")
axs[0, 0].legend()

axs[0, 1].plot(x, np.cos(x), label='cos(x)', color='green')
axs[0, 1].set_title("Cosine")
axs[0, 1].legend()

axs[1, 0].plot(x, np.exp(-x), label='exp(-x)', color='purple')
axs[1, 0].set_title("Exponential Decay")
axs[1, 0].legend()

axs[1, 1].plot(x, np.log(x + 1), label='log(x+1)', color='orange')
axs[1, 1].set_title("Logarithm")
axs[1, 1].legend()

# Add common labels
for ax in axs.flat:
    ax.set(xlabel='x', ylabel='f(x)')
    ax.grid(True)

# Avoid overlap
plt.tight_layout()
plt.show()
````

## Summary: What You Can Now Do

| Skill                        | Tool                        |
|-----------------------------|-----------------------------|
| Create subplot grids        | `plt.subplots(rows, cols)`  |
| Share axes                  | `sharex`, `sharey`          |
| Access individual subplots  | `axs[row, col]`             |
| Add spacing                 | `plt.tight_layout()`        |
| Customize everything        | `ax.set_*` methods          |
