## Getting Started with SciPy

SciPy is built on top of NumPy and is organized into sub-packages such as `scipy.integrate`, `scipy.optimize`, `scipy.linalg`, and `scipy.stats`.

Let's install and import the library first:


```python
# Uncomment if not already installed
# !pip install scipy

import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate, optimize, linalg, stats, signal, interpolate
```

## Optimization

Now let's find the **minimum** of a nonlinear function:

$
f(x) = x^2 + 10 \sin(x)
$

This is a classic example used to test optimization algorithms.


```python
fun = lambda x: x**2 + 10 * np.sin(x)
res = optimize.minimize(fun, x0=2)
print(f"Minimum occurs at x = {res.x[0]:.4f}, f(x) = {res.fun:.4f}")
```

## Linear Algebra

Let’s solve a linear system of equations:

$
\begin{bmatrix}
3 & 2 \\
1 & 4
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
5 \\
6
\end{bmatrix}
$


```python
A = np.array([[3, 2], [1, 4]])
b = np.array([5, 6])
solution = linalg.solve(A, b)
print(f"Solution: x = {solution[0]:.2f}, y = {solution[1]:.2f}")
```

## Statistical Analysis

SciPy's `stats` module offers a vast set of tools for statistics. Let’s:

- Generate 1000 random samples from a normal distribution
- Plot the histogram
- Test for normality

**In normality tests, the null hypothesis (H0) states that the data comes from a normally distributed population, meaning the data are well-modeled by a normal distribution. If a normality test shows a low p-value (typically less than 0.05), the null hypothesis is rejected, indicating that the data deviate significantly from a normal distribution. If the p-value is high, the null hypothesis is not rejected, suggesting that the data can be reasonably assumed to be normally distributed.**


```python
data = stats.norm.rvs(loc=0, scale=1, size=1000)
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
plt.title("Histogram of Normally Distributed Data")
plt.show()

# Perform a normality test
k2, p = stats.normaltest(data)
print(f"p-value = {p:.4f}")
```

## Signal Processing

We’ll filter a noisy signal using a **Butterworth low-pass filter**.


```python
t = np.linspace(0, 1.0, 200)
x = np.sin(2*np.pi*5*t) + 0.5 * np.random.randn(200)

# Design Butterworth filter
b, a = signal.butter(3, 0.1)
filtered = signal.filtfilt(b, a, x)

plt.plot(t, x, label='Noisy Signal')
plt.plot(t, filtered, label='Filtered Signal', linewidth=2)
plt.legend()
plt.title("Signal Filtering with SciPy")
plt.show()
```

## Interpolation

We’ll demonstrate interpolation on a noisy sine curve using `scipy.interpolate`.


```python
x = np.linspace(0, 10, 10)
y = np.sin(x) + 0.2 * np.random.randn(10)

f_interp = interpolate.interp1d(x, y, kind='cubic')
x_new = np.linspace(0, 10, 100)
y_new = f_interp(x_new)

plt.plot(x, y, 'o', label='Data')
plt.plot(x_new, y_new, '-', label='Cubic Interpolation')
plt.legend()
plt.title("Interpolation Example")
plt.show()
```
