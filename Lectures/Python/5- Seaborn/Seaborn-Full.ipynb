# Exploring Distributions

Distribution plots help you understand how values of a numeric variable are spread out. Seaborn provides multiple tools for visualizing distributions:

- `displot()` — combined histogram and KDE for univariate data
- `histplot()` — just the histogram
- `kdeplot()` — kernel density estimation
- `rugplot()` — marks individual data points
- `jointplot()` — shows bivariate distributions
- `pairplot()` — quick overview of numeric relationships in a dataset

We'll use the `penguins` dataset for this section.


````python
#imports

import seaborn as sns
import matplotlib.pyplot as plt


# Load data
penguins = sns.load_dataset('penguins')
````

````python
penguins
````

## Histogram + KDE

````python

sns.displot(penguins, x='flipper_length_mm', kde=True)
plt.title('Flipper Length Distribution')
plt.show()
````

## Bivariate relationships

````python
sns.jointplot(data=penguins, x='flipper_length_mm', y='body_mass_g', kind='scatter')
````

## Pairwise visualizations


````python
sns.pairplot(penguins, hue='species')
````

## KDE + Rugplot


````python
fig,ax=plt.subplots(figsize=(10,6))
sns.kdeplot(data=penguins, x='bill_length_mm')
sns.rugplot(data=penguins, x='bill_length_mm')
plt.title('KDE and Rugplot of Bill Length')
plt.show()
````

# Visualizing Categorical Data

Seaborn provides a variety of plots for summarizing data grouped by categorical variables:

- `barplot()` — shows the central tendency
- `countplot()` — counts occurrences
- `boxplot()` — displays quartiles and outliers
- `violinplot()` — combines KDE with a boxplot
- `stripplot()` and `swarmplot()` — shows all points with or without overlap

We'll use the `titanic` dataset in this section.


````python
# Load dataset
titanic = sns.load_dataset('titanic')
````

````python
titanic.head()
````

## Barplot and countplos

````python
fig,ax=plt.subplots(figsize=(10,6))
sns.barplot(data=titanic, x='sex', y='age', estimator=lambda x: x.mean())
sns.countplot(data=titanic, x='class')
plt.show()
````

## Box and violin plots

````python
fig,ax=plt.subplots(figsize=(10,6))
sns.boxplot(data=titanic, x='class', y='age', palette='Set3')
sns.violinplot(data=titanic, x='class', y='age', hue='sex', split=True)
plt.show()
````

## Stripplot


````python
sns.stripplot(data=titanic, x='class', y='fare', jitter=True, hue='sex', dodge=True)
plt.show()
````

# Section 3: Heatmaps and Cluster Maps

Matrix plots help you visualize numeric relationships using color intensities.

- `heatmap()` — shows correlation or pivot tables
- `clustermap()` — adds clustering on top of a heatmap

We'll use the `flights` dataset to demonstrate.


````python
# Load and transform dataset

flights = sns.load_dataset('flights')
pivot_flights = flights.pivot_table(index='month', columns='year', values='passengers')
````

````python
pivot_flights
````

## Heatmap


````python
fig,ax=plt.subplots(figsize=(20,6))

sns.heatmap(pivot_flights, cmap='YlGnBu', annot=False)
plt.title('Flight Passengers Heatmap')
plt.show()
````

# Section 4: Using Grids for Multi-Plot Layouts

Seaborn provides flexible tools to generate multiple plots categorized by features:

- `FacetGrid()` — apply the same plot across subsets
- `PairGrid()` — manually build pairwise plots
- `JointGrid()` — fully custom joint plot

Here, we’ll work with the `exercise` and `iris` datasets.


## PairGrid example (iris)


````python
iris = sns.load_dataset('iris')
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, fill=True)
````

## FacetGrid example

````python
# FacetGrid example (exercise)
exercise = sns.load_dataset('exercise')
g = sns.FacetGrid(exercise, col="diet", row="time")
g.map_dataframe(sns.barplot, x="kind", y="pulse")
````

## JointGrid example (penguins)


````python
g = sns.JointGrid(data=penguins, x='flipper_length_mm', y='body_mass_g')
g.plot(sns.scatterplot, sns.histplot)
````

# Section 5: Regression-Based Visualizations

Seaborn makes it simple to overlay linear regression fits:

- `regplot()` — low-level regression plot

We'll use `diamonds` dataset for this section.


````python
diamonds = sns.load_dataset('diamonds')

# Basic regression
sns.lmplot(data=diamonds.sample(500), x='carat', y='price')
````
